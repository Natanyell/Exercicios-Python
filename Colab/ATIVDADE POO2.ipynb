```python
class Cliente:
    def __init__(self, nome, ID, idade, DataNascimento):
        self.nome = nome
        self.ID = ID
        self.idade = idade
        self.DataNascimento = DataNascimento

    def get_nome(self):
        return self.nome

    def set_nome(self, nome):
        self.nome = nome

    def get_ID(self):
        return self.ID

    def set_ID(self, ID):
        self.ID = ID
        

    def get_idade(self):
        return self.idade

    def set_idade(self, idade):
        self.idade = idade

    def get_DataNascimento(self):
        return self.DataNascimento

    def set_DataNascimento(self, DataNascimento):
        self.DataNascimento = DataNascimento
```

```python
class Livro:
    def __init__(self, titulo, autor, id):
        self.titulo = titulo
        self.autor = autor
        self.id = id

    def get_titulo(self):
        return self.titulo

    def set_titulo(self, titulo):
        self.titulo = titulo

    def get_autor(self):
        return self.autor

    def set_autor(self, autor):
        self.autor = autor

    def get_id(self):
        return self.id

    def set_id(self, id):
        self.id = id
```

```python
# Criando instâncias das classes
cliente = Cliente("Nome", 'id', "idade", "data de nascimento")
livro1 = Livro("nome do livro", "autor", 'id')


# Definindo novos valores para alguns atributos das instâncias usando os métodos set
cliente.set_nome(input("Digite seu nome: "))
cliente.set_ID(float(input('digite o id do cliente: ')))
cliente.set_idade(int(input(("Digite sua idade: "))))
cliente.set_DataNascimento(input("Digite Sua Data De Nascimento: "))
livro1.set_titulo(input("Digite o titulo do livro: "))
livro1.set_autor(input("Digite o nome do autor: "))
livro1.set_id(input("Digite o id do livro: "))

# Obtendo e imprimindo atributos das instâncias
print()
print()
print("-"*20)
print()
print()
print("Nome do cliente: ", cliente.get_nome())
print("ID do cliente: ", cliente.get_ID())
print("Idade do cliente: ", cliente.get_idade())
print("Data de nascimento do cliente: ", cliente.get_DataNascimento())
print("Título do livro: ", livro1.get_titulo())
print("Autor do livro: ", livro1.get_autor())
print("ID do livro: ", livro1.get_id())
```
