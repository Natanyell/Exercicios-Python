```python
import random
import ipywidgets as widgets
from IPython.display import display

def move_button_1(x, y):
    if abs(x - button_1_x) < 50 and abs(y - button_1_y) < 50:
        button_1.layout.left = f'{random.randint(0, colab_width - button_1_width)}px'
        button_1.layout.top = f'{random.randint(0, colab_height - button_1_height)}px'

def accepted(b):
    display(widgets.HTML('<b>Eu te amo meu amor, lanchinho mais tarde?</b>'))

def denied(b):
    button_1.close()

button_1_x = -100
button_1_y = -100
button_1_width = 100
button_1_height = 30
colab_width = 800
colab_height = 400

button_1 = widgets.Button(description='Não')
button_1.layout.position = 'absolute'
button_1.layout.left = f'{button_1_x}px'
button_1.layout.top = f'{button_1_y}px'
button_1.on_click(denied)

button_2 = widgets.Button(description='Sim')
button_2.layout.position = 'absolute'
button_2.layout.left = '10px'
button_2.layout.top = '10px'
button_2.on_click(accepted)

display(button_1)
display(button_2)

widgets.interact(move_button_1, x=widgets.IntSlider(min=0, max=colab_width, continuous_update=False),
                 y=widgets.IntSlider(min=0, max=colab_height, continuous_update=False))
```
