```python
import random
from IPython.display import display, Javascript

def move_button_1(x, y):
    if abs(x - button_1_x) < 50 and abs(y - button_1_y) < 50:
        new_x = random.randint(0, colab_width - button_1_width)
        new_y = random.randint(0, colab_height - button_1_height)
        button_1_js = f"document.getElementById('button-1').style.left = '{new_x}px';"
        button_1_js += f"document.getElementById('button-1').style.top = '{new_y}px';"
        display(Javascript(button_1_js))

def accepted():
    display(Javascript("alert('Eu te amo meu amor, lanchinho mais tarde?')"))

def denied():
    display(Javascript("document.getElementById('button-1').remove();"))

button_1_x = -100
button_1_y = -100
button_1_width = 100
button_1_height = 30
colab_width = 800
colab_height = 400

button_1_js = '''
var button = document.createElement('button');
button.id = 'button-1';
button.style.position = 'absolute';
button.style.left = '-100px';
button.style.top = '-100px';
button.innerText = 'Não';
button.onclick = denied;
document.body.appendChild(button);
'''

button_2_js = '''
var button = document.createElement('button');
button.id = 'button-2';
button.style.position = 'absolute';
button.style.left = '10px';
button.style.top = '10px';
button.innerText = 'Sim';
button.onclick = accepted;
document.body.appendChild(button);
'''

display(Javascript(button_1_js))
display(Javascript(button_2_js))
display(Javascript('''
document.addEventListener('mousemove', function(e) {
    var x = e.clientX || e.pageX;
    var y = e.clientY || e.pageY;
    google.colab.kernel.invokeFunction('notebook.move_button_1', [x, y], {});
});
'''))
```
