```python
class VeiculoParticular:
    def __init__(self, Id, placa, marca, ano):
        self.Id = Id
        self.placa = placa
        self.marca = marca
        self.ano = ano

    def get_Id(self):
        return self.Id

    def set_Id(self, Id):
        self.Id = Id

    def get_placa(self):
        return self.placa

    def set_placa(self, placa):
        self.placa = placa

    def get_marca(self):
        return self.marca

    def set_marca(self, marca):
        self.marca = marca

    def get_ano(self):
        return self.ano

    def set_ano(self, ano):
        self.ano = ano
```

```python
class VeiculoColetivo:
    def __init__(self, Id, placa, marca, QtdEixos):
        self.Id = Id
        self.placa = placa
        self.marca = marca
        self.QtdEixos = QtdEixos

    def get_Id(self):
        return self.Id

    def set_Id(self, Id):
        self.Id = Id

    def get_placa(self):
        return self.placa

    def set_placa(self, placa):
        self.placa = placa

    def get_marca(self):
        return self.marca

    def set_marca(self, marca):
        self.marca = marca

    def get_QtdEixos(self):
        return self.QtdEixos

    def set_QtdEixos(self, QtdEixos):
        self.QtdEixos = QtdEixos
```

```python
# Criando instâncias das classes

veiculo1 = VeiculoParticular('Id', 'placa', 'marca', 'ano')
veiculo2 = VeiculoColetivo('Id', 'placa', 'marca', 'QtdEixos')

#Definindo novos valores para alguns atributos das instâncias usando os métodos set

print('Veiculo Particular')
print()

veiculo1.set_Id(float(input("Digite o ID: ")))
veiculo1.set_placa(input("Digite a Placa do veiculo particular: "))
veiculo1.set_marca(input("Digite a Marca Do Veiculo particular: "))
veiculo1.set_ano(float(input("Digite o Ano Do Veiculo particular: ")))

print()
print('Veiculo Coletivo')
print()

veiculo2.set_Id(float(input("Digite o ID: ")))
veiculo2.set_placa(input("Digite a Placa do veiculo coletivo: "))
veiculo2.set_marca(input("Digite a Marca do veiculo coletivo: "))
veiculo2.set_QtdEixos(int(input("Digite a quantidade de eixos do veiculo coletivo: ")))

print()

print("Espere 5 segundos, estamos analisando..." )

import time
time.sleep(5)

print()
#Recebo da impressão os atributos que você dá as instâncias 
print("="  *  40)   
print() 
print('Veiculo Particular')   
print() 
print("ID: ", veiculo1.get_Id())   
print("Placa Do Veículo: ", veiculo1.get_placa())  
print("Marca Do Veículo: ", veiculo1.get_marca())   
print("Ano Do Veiculo: ", veiculo1.get_ano())
print()
print('Veiculo Coletivo')
print()
print("ID: ", veiculo2.get_Id())
print("Placa Do Veículo: ", veiculo2.get_placa())
print("Marca Do Veiculo: ", veiculo2.get_marca())
print("Quantidade De Eixos: ", veiculo2.get_QtdEixos())
```
